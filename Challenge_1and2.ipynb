Note that there is negative pax. That corresponds to cancelations. So to get the total number of passengers that have actually booked, you should sum including the negatives (that will remove the canceled bookings).

<font color = '#1ABC9C'> For process all data we will apply chunk.
 I need know how many rows are in the data set for solved the exercise 1. I add loops for calculate this in the chunks.

```python
import pandas as pd

pd.set_option('display.max_columns', None)
cols = ['arr_port', 'year', 'pax']
chunk = 300000
bookings = pd.read_csv('~/Data/challenge/bookings.csv', sep = '^', usecols = cols, chunksize = chunk)

#We create a list of chunk empty for then we link all chunks generate in the procces:
rows = 0 #counter of rows in dataset(for chunks)
chunks = 0 #number of chunks in the process
all_chunks = pd.DataFrame()

for chunk in bookings:
    bookings_chunk = chunk.dropna()
    bookings_chunk = chunk[chunk['year'] == 2013]
    bookings_chunk_group = bookings_chunk.groupby('arr_port')['pax'].agg('sum').reset_index()
    
    
    for i, r in enumerate(bookings_chunk_group['arr_port']):
        r1 = r.strip()
        bookings_chunk_group.at[i, 'arr_port'] = r1
    
    chunks += 1
    rows += len(bookings_chunk)
    
    all_chunks = all_chunks.append(bookings_chunk_group)
    
    all_chunks = all_chunks.groupby('arr_port')['pax'].agg('sum').reset_index()

top_10_arr_port_for_pax = all_chunks.sort_values('pax', ascending = False ).head(10)
top_10_arr_port_for_pax
print('Total number of chunks in the dataset are: %d and rows are: %d'%(chunks, rows))
```
