```python
# Process by chunks

#Me da error de Date a partir de la 1.436.000

cols = ['Date', 'Origin', 'Destination'] 
filter_destination = ['AGP', 'MAD', 'BCN']

all_chunks = pd.DataFrame()

rows = 0
chunks = 0

for chunk in pd.read_csv("~/Data/challenge/searches.csv",sep = '^', usecols = cols, chunksize = 500000):
    
    chunk = chunk.dropna()
    rows += len(chunk)
    chunks += 1
    
    chunk = chunk[chunk.Destination.isin(filter_destination)]
    chunk['counter'] = 1
   
    group_destination = chunk.groupby(['Destination', 'Date'])['counter'].sum().reset_index()
    
    group_destination['Date'] = pd.to_datetime(group_destination['Date'])
    group_destination['month'] = group_destination['Date'].dt.month
    group_destination['year'] = group_destination['Date'].dt.year
    
    group_destination = group_destination.groupby(['Destination', 'month'])['counter'].sum().reset_index()
    
    group_destination = group_destination.sort_values(by = 'counter', ascending = False)
    
    
    
    print(str(rows))
    all_chunks = all_chunks.append(group_destination)
    
    all_chunks = all_chunks.groupby(['Destination', 'month'])['counter'].agg('sum').reset_index()
    
    all_chunks
print(str(rows))
print(str(chunks))
print(all_chunks)
```
