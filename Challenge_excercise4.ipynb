```python
import pandas as pd

pd.set_option('display.max_columns', None)
cols = ['arr_port','dep_port', 'year', 'pax']
chunk = 500000
bookings = pd.read_csv('~/Data/challenge/bookings.csv', sep = '^', usecols = cols, chunksize = chunk)

#We create a list of chunk empty for then we link all chunks generate in the procces:
rows = 0 #counter of rows in dataset(for chunks)
chunks = 0 #number of chunks in the process
bookings_ok = pd.DataFrame()

for chunk in bookings:
    bookings_chunk = chunk.dropna()
    bookings_chunk = chunk[chunk['year'] == 2013]
    booking_chunk = chunk[chunk['pax'] >= 1]
    
    for i, r in enumerate(bookings_chunk['arr_port']):
        r1 = r.strip()
        bookings_chunk.at[i, 'arr_port'] = r1
    
    for i, r in enumerate(bookings_chunk['dep_port']):
        r2 = r.strip()
        bookings_chunk.at[i, 'dep_port'] = r2
    
    chunks += 1
    rows += len(bookings_chunk)
    
    bookings_chunk['Common_Column'] = bookings_chunk['act_date'].astype(str) + ' ' + bookings_chunk['dep_port'] + ' ' + bookings_chunk['arr_port']
    
    bookings_chunk['Counter'] = 1
    
    bookings_ok = bookings_ok.append(bookings_chunk)
    
    
    
bookings_ok.to_csv('Bookings_ok.csv')
```
